```python
from typing import Text

w = widgets.Dropdown(
    options=['Addition', 'Multiplication', 'Subtraction', 'Division'],
    value='Addition',
    description='Task:',
)
txt1=widgets.Text()
txt1.value="1"
display(txt1)

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        print("changed to %s" % change['new'])
    txt1.value=w.value

w.observe(on_change)

display(w)
```
